```python
import pandas as pd
from functools import reduce
```

```python
files = ["transactions_2025_01_with_category.xlsx",
         "transactions_2025_02_with_category.xlsx",
         "transactions_2025_03_with_category.xlsx",
        "transactions_2025_04_with_category.xlsx",
        "transactions_2025_05_with_category.xlsx",
        "transactions_2025_06_with_category.xlsx",
        "transactions_2025_07_with_category.xlsx",
        "transactions_2025_08_with_category.xlsx",
        "transactions_2025_09_with_category.xlsx",
        "transactions_2025_10_with_category.xlsx",
        "transactions_2025_11_with_category.xlsx",
        "transactions_2025_12_with_category.xlsx",
        "transactions_2026_01_with_category.xlsx",
        "transactions_2026_02_with_category.xlsx"]
```

```python
#Read all Excel files
dfs = [pd.read_excel(file) for file in files]

#common columns across all datasets
common_columns = set(dfs[0].columns)
for df in dfs[1:]:
    common_columns &= set(df.columns)
common_columns = list(common_columns)
dfs_common = [df[common_columns] for df in dfs]
```

```python
combined_df = pd.concat(dfs_common,ignore_index=True)

combined_df.to_csv("combined_dataset.csv",index=False)
```
