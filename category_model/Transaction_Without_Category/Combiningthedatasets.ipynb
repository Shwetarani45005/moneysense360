```python
import pandas as pd
from functools import reduce
```

```python
files = ["transactions_2025_01_no_category.xlsx",
         "transactions_2025_02_no_category.xlsx",
         "transactions_2025_03_no_category.xlsx",
        "transactions_2025_04_no_category.xlsx",
        "transactions_2025_05_no_category.xlsx",
        "transactions_2025_06_no_category.xlsx",
        "transactions_2025_07_no_category.xlsx",
        "transactions_2025_08_no_category.xlsx",
        "transactions_2025_09_no_category.xlsx",
        "transactions_2025_10_no_category.xlsx",
        "transactions_2025_11_no_category.xlsx",
        "transactions_2025_12_no_category.xlsx",
        "transactions_2026_01_no_category.xlsx",
        "transactions_2026_02_no_category.xlsx"]
```

```python
#Read all Excel files
dfs = [pd.read_excel(file) for file in files]
```

```python
#common columns across all datasets
common_columns = set(dfs[0].columns)
for df in dfs[1:]:
    common_columns &= set(df.columns)
common_columns = list(common_columns)
dfs_common = [df[common_columns] for df in dfs]
```

```python
combined_df = pd.concat(dfs_common,ignore_index=True)

combined_df.to_csv("combined_dataset.csv",index=False)
```
